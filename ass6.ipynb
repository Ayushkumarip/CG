```python
import pandas as pd

data = {
    "time": ["10:00","10:01","10:02","10:03","10:04"],
    "#AI": [5,7,9,6,8],
    "#Spark": [3,4,5,7,6],
    "#ML": [2,3,4,5,6]
}

df = pd.DataFrame(data)
```

```python
import plotly.express as px

fig = px.line(df, x="time", y=["#AI","#Spark","#ML"],
              title="Hashtag Trends Over Time")

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Bar(x=df["time"], y=df["#AI"], name="#AI"))
fig.add_trace(go.Bar(x=df["time"], y=df["#Spark"], name="#Spark"))

fig.update_layout(title="Real-time Hashtag Dashboard")

fig.show()
```

```python
import plotly.express as px

geo_data = pd.DataFrame({
    "lat": [19.07, 19.08, 19.09],
    "lon": [72.87, 72.88, 72.86],
    "count": [10, 20, 15]
})

fig = px.scatter_mapbox(
    geo_data,
    lat="lat",
    lon="lon",
    size="count",
    color="count",
    mapbox_style="open-street-map",
    title="Geographic Heatmap"
)

fig.show()
```

```python
fig.write_html("dashboard.html")
```
